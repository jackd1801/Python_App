```python
import pandas as pd 
import numpy as np
```

```python
url = 'Data/ukpopestimatesmid2021on2021geographyfinal.xls'
df = pd.read_excel(url, sheet_name='MYE2 - Persons', usecols='A:CQ', skiprows=range(0, 7), nrows=428-7)

df_melted = df.melt(id_vars=['Code', 'Name', 'Geography', 'All ages'], value_vars=list(map(str, range(0, 90))) + ['90+'], 
                    var_name='Age', value_name='Population')
```

```python
shapefilela = pd.read_csv("Data/Middle_Layer_Super_Output_Areas_(Dec_2021)_Boundaries_Full_Clipped_EW_(BFC).csv")

df_joined = df_melted.merge(shapefilela, how="left", left_on="Code", right_on="MSOA21CD")

df_joined
```
